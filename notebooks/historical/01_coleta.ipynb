```python
import requests
import zipfile
import io
import pandas as pd
from datetime import datetime, timedelta
import os
```

### Divisao de diretorios para live e historico

```python
os.makedirs("data/raw/historical", exist_ok=True)
```

### Funcao extracao

```python

def baixar_gdelt_2(timestamp):
    ts_str = timestamp.strftime("%Y%m%d%H%M%S")
    url = f"http://data.gdeltproject.org/gdeltv2/{ts_str}.export.CSV.zip"

    try:
        response = requests.get(url, timeout=10)

        if response.status_code != 200:
            print(f"[ERRO] {ts_str}")
            return None

        z = zipfile.ZipFile(io.BytesIO(response.content))
        csv_file = z.open(z.namelist()[0])

        df = pd.read_csv(csv_file, sep="\t", header=None)

        return df

    except Exception as e:
        print(f"[EXCEPTION] {ts_str} -> {e}")
        return None
```

```python
def download_historico(start_date, end_date, output_path="data/raw/historical"):
    os.makedirs(output_path, exist_ok=True)

    current = start_date
    buffer = []

    while current <= end_date:
        print(f"[HIST] {current}")

        df = baixar_gdelt_2(current)

        if df is not None:
            buffer.append(df)

        if current.hour == 23 and current.minute == 45:
            if buffer:
                df_day = pd.concat(buffer)

                date_str = current.strftime("%Y%m%d")
                df_day.to_parquet(f"{output_path}/{date_str}.parquet")

                print(f"[SALVO DIA] {date_str}")

                buffer = []

        current += timedelta(minutes=15)
```

```python
start = datetime(2025, 1, 1) 
end   = datetime(2025, 1, 7)

download_historico(start, end)
```
