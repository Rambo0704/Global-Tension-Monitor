```python
import pandas as pd
import os
```

```python
PROCESSED_DIR = "C:/Users/Gabriel/Documents/Projects/Global-Tension-Monitor/data/processed"
os.makedirs(PROCESSED_DIR, exist_ok=True)
```

```python
PROJECT_ID = "global-tension-monitor"
```

```python
query = """
    SELECT
      PARSE_DATE('%Y%m%d', CAST(SQLDATE AS STRING)) AS date,
      Actor1CountryCode AS actor1_country,
      AVG(GoldsteinScale) AS goldstein,
      AVG(AvgTone) AS tone,
      SUM(NumMentions) AS mentions,
      SUM(NumSources) AS sources,
      SUM(NumArticles) AS articles
    FROM
      `gdelt-bq.gdeltv2.events`
    WHERE
      SQLDATE >= 20160101 AND SQLDATE <= 20260101
      AND Actor1CountryCode IS NOT NULL
    GROUP BY
      date,
      actor1_country
    ORDER BY
      date,
      actor1_country
"""

print("Enviando query para os servidores do Google ")
```

```python

df_historico = pd.read_gbq(query, project_id=PROJECT_ID, dialect='standard')


df_historico["date"] = pd.to_datetime(df_historico["date"], format="%Y-%m-%d", errors="coerce")


df_historico["goldstein"] = pd.to_numeric(df_historico["goldstein"], errors="coerce")
df_historico["tone"] = pd.to_numeric(df_historico["tone"], errors="coerce")
df_historico["mentions"] = pd.to_numeric(df_historico["mentions"], errors="coerce")
df_historico["sources"] = pd.to_numeric(df_historico["sources"], errors="coerce")
df_historico["articles"] = pd.to_numeric(df_historico["articles"], errors="coerce")
```

```python

output_file = os.path.join(PROCESSED_DIR, "historico_agrupado_10anos.parquet")
df_historico.to_parquet(output_file, index=False)

print(f"[SUCESSO] Base extraída e salva em: {output_file}")
```
