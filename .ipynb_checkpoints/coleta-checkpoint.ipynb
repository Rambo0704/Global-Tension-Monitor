```python
import requests
import zipfile
import io
import pandas as pd
from datetime import datetime, timedelta
import os
```

```python
os.makedirs("data/raw/historical", exist_ok=True)
os.makedirs("data/raw/live", exist_ok=True)
```

```python
def baixar_dia(date):
    date_str = date.strftime("%Y%m%d")
    url = f"http://data.gdeltproject.org/events/{date_str}.export.CSV.zip"

    try:
        response = requests.get(url)
        z = zipfile.ZipFile(io.BytesIO(response.content))
        csv_file = z.open(z.namelist()[0])

        df = pd.read_csv(csv_file, sep="\t", header=None, dtype=str)
        return df

    except Exception as e:
        print(f"Erro ao baixar {date_str}: {e}")
        return None
```

```python
start = datetime(2020, 1, 1)
end   = datetime(2020, 3, 1)

date = start
while date <= end:
    print(f"[HIST] Baixando {date}")
    
    df = baixar_dia(date)

    if df is not None:
        date_str = date.strftime("%Y%m%d")
        df.to_parquet(f"data/raw/historical/{date_str}.parquet")

    date += timedelta(days=1)
```

```python
end = datetime.today()
start = end - timedelta(days=30)

date = start
while date <= end:
    print(f"[LIVE] Baixando {date}")
    
    df = baixar_dia(date)

    if df is not None:
        date_str = date.strftime("%Y%m%d")
        df.to_parquet(f"data/raw/live/{date_str}.parquet")

    date += timedelta(days=1)
```
